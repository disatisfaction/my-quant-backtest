```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
HS300constituentdaily = pd.read_csv("HS300constituentdaily.csv",index_col=[0,1])
dates = HS300constituentdaily.index.levels[0]
HS300constituentdaily
```

```python
data_open = HS300constituentdaily["open"].unstack()
data_open
```

```python
# 股票每单位时间的收益率数据框，列名代表每支股票，索引代表依次的每个时间点
data_return = data_open / data_open.shift() - 1
data_return
```

```python
# 股票每次需要换仓时各个股票的仓位，即投资该股票的资金占头寸的比例，1减去所有股票的仓位之和即为保留的现金占头寸的比例，索引代表分别在第几个时间点换仓
pos_at_chg_time_point = pd.read_csv('pos_at_chg_time_point.csv', index_col=0)
pos_at_chg_time_point
```

```python
buy_fee_rate = 0.00025
sell_fee_rate = 0.00125
```

```python
# 使用股票每次需要换仓时各个股票的仓位的数据框代表一个择股策略，依据股票收益率数据，以下函数可计算投资组合的净值变化（pnl）
# 为计算简单起见，不考虑股票份数的整数性要求，
# 且认为换仓时一次交易同时完成所有需要的买入卖出，完成交易后所有买入卖出手续费同时直接作用在投资组合的净值上，而非使用现金支付，该假设使得可以先在不考虑手续费的情况下完成投资组合净值变化的计算，最后再在每个换仓时间点，让它后面的所有时间点的净值乘上这次换仓手续费造成的影响
def my_backtest(
    data_return: pd.DataFrame,
    pos_at_chg_time_point: pd.DataFrame,
    buy_fee_rate: float = 0,
    sell_fee_rate: float = 0,
) -> pd.Series:
    # 参数：
    # data_return: 股票收益率数据，DataFrame，index具体时间（日期），columns为股票代码
    # pos_at_chg_time_point: 每次换仓时各个股票的仓位，DataFrame，index为换仓时间点，columns为股票代码，可为data_return的columns的子集
    # buy_fee_rate: 买入费率，float
    # sell_fee_rate: 卖出费率，float
    # 返回值：
    # net_value: 考虑手续费后，投资组合每个时间点的净值，Series=

    n = len(data_return.index)  # 时间点个数
    m = len(data_return.columns)  # 股票数
    m_involved = len(pos_at_chg_time_point.columns)  # 涉及的股票数
    k = len(pos_at_chg_time_point)  # 换仓时间点数

    # pos_at_chg_time_point = pos_at_chg_time_point.reindex(columns=data_return.columns, fill_value=0)
    data_return = data_return[pos_at_chg_time_point.columns]

    pos_chg_time_point = pos_at_chg_time_point.index  # 换仓时间点

    repeats = np.diff(
        np.concatenate([[0], pos_chg_time_point + 1, [n]])
    )  # 原始空仓和每一次换仓后持续未换仓的时间点个数，不计开始时间点，计结束时间点

    reference_pos = pd.DataFrame(
        np.repeat(
            np.vstack([[[0] * m_involved], pos_at_chg_time_point]),
            repeats=repeats,
            axis=0,
        ),
        index=data_return.index,
        columns=pos_at_chg_time_point.columns,
    )  # 每只股票每个时间点对应的上一换仓时间点的仓位（第一次换仓的上一换仓为原始空仓）

    data_whole_return_compared_to_reference = (
        (data_return.fillna(0) + 1)
        .groupby(np.repeat(np.arange(-1, k), repeats))
        .cumprod()
    )  # 每只股票每个时间点相较对应的上一换仓时间点的全收益率（收益率加1）

    reference_inv = reference_pos.sum(
        axis=1
    )  # 每个时间点对应的上一换仓时间点的股票总投资仓位

    reference_cash = 1 - reference_inv  # 每个时间点对应的上一换仓时间点的现金仓位

    pos_compared_to_reference = (
        reference_pos * data_whole_return_compared_to_reference
    )  # 每只股票每个时间点，以上一换仓时间点的投资组合净值为1，此时的净值

    whole_return_compared_to_reference = (
        pos_compared_to_reference.sum(axis=1) + reference_cash
    )  # 投资组合每个时间点相较对应的上一换仓时间的全收益率

    whole_cum_return_at_chg_time_point = whole_return_compared_to_reference.iloc[
        pos_chg_time_point
    ].cumprod()  # 每个换仓时间点的投资组合净值，即相较原始净值的全收益率

    reference_whole_cum_return = np.repeat(
        np.concatenate([[1], whole_cum_return_at_chg_time_point]),
        repeats=repeats,
    )  # 投资组合每个时间点对应的上一换仓时间点的投资组合净值

    whole_cum_return = (
        whole_return_compared_to_reference * reference_whole_cum_return
    )  # 投资组合每个时间点的净值


    pos_compared_to_reference_at_chg_time_point = pos_compared_to_reference.iloc[
        pos_chg_time_point
    ]
    asset_compared_to_reference_at_chg_time_point = (
        pos_compared_to_reference_at_chg_time_point.sum(axis=1)
        + reference_cash.iloc[pos_chg_time_point]
    )  # 投资组合每个换仓时间点，以上一换仓时间点的投资组合净值为1，此时的净值

    pos_chg_at_chg_time_point = (
        pos_at_chg_time_point
        - (
            pos_compared_to_reference_at_chg_time_point.div(
                asset_compared_to_reference_at_chg_time_point, axis=0
            )
        ).values
    )  # 投资组合每个换仓时间点，换仓导致的每只股票的仓位变化

    fee_rate_at_chg_time_point = (
        pos_chg_at_chg_time_point
        * np.where(pos_chg_at_chg_time_point > 0, buy_fee_rate, -sell_fee_rate)
    ).sum(
        axis=1
    )  # 投资组合每个换仓时间点，换仓导致的手续费比率
    repeats2 = np.concatenate(
        [[repeats[0] - 1], repeats[1:-1], [repeats[-1] + 1]]
    )  # 原始空仓和每一次换仓后持续未换仓的时间点个数，计开始时间点，不计结束时间点
    cum_left_from_fee_rate_at_chg_time_point = (
        1 - fee_rate_at_chg_time_point
    ).cumprod()  # 投资组合每个换仓时间点，换仓导致的手续费导致的净值缩小比率的累计
    cum_left_from_fee_rate = np.repeat(
        np.concatenate([[1], cum_left_from_fee_rate_at_chg_time_point]),
        repeats=repeats2,
    )  # 投资组合每个时间点，已经历的换仓导致的手续费导致的净值缩小比率的累计

    net_value = (
        whole_cum_return * cum_left_from_fee_rate
    )  # 考虑手续费后，投资组合每个时间点的净值

    return net_value
```

```python
def max_draw_down(net_value):
    cum_max_net_value = np.maximum.accumulate(net_value)
    end = np.argmax((cum_max_net_value - net_value) / cum_max_net_value) # 结束位置
    begin = np.argmax(net_value[:end]) # 开始位置
    return net_value[end] - net_value[begin], begin, end
```

```python
net_value = my_backtest(data_return=data_return, pos_at_chg_time_point=pos_at_chg_time_point, buy_fee_rate=buy_fee_rate, sell_fee_rate=sell_fee_rate)
```
